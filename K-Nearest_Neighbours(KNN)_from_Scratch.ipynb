```python
import numpy as np
from scipy.stats import mode
 
def eucledian_distance(p1,p2):
    '''
    Calculating Eucledian distance.
    
    Args:
    p1, p2: Co-ordinates of the points.
    
    Returns:
    dist: Distance (float)
    '''
    dist = np.sqrt(np.sum((p1-p2)**2))
    return dist
 

def predict(x_train, y , x_input, k):
    '''
    Function to calculate KNN
    
    Args:
    x_train: np.array of features for tarining.
    y: np.array of labels for tarining.
    x_input: np.array of features for prediction.
    k: No of neighboures (int)
    
    '''
    op_labels = []
     
    #Looping through the Datapoints to be classified
    for item in x_input: 
         
        #Array to store distances
        point_dist = []
         
        #Looping through each training Data
        for j in range(len(x_train)): 
            distances = eucledian_distance(np.array(x_train[j,:]) , item) 
            #Calculating the distances
            point_dist.append(distances) 
        point_dist = np.array(point_dist) 
         
        #Sorting the array while preserving the index
        dist = np.argsort(point_dist)[:k] 
         
        #Labels of the K datapoint
        labels = y[dist]
         
        #Voting for majority
        lab = mode(labels) 
        lab = lab[0]
        op_labels.append(lab)
 
    return op_labels
```

```python
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from numpy.random import randint
import random

#Loading the Data from sklearn
iris= load_iris()

#Store features matrix in X
X= iris.data
#Store target vector in y
y= iris.target

#Creating the training Dataset
data_index = [i for i in range(150)]
random.shuffle(data_index)
#train_randint = randint(0,150,100)
X_train = X[data_index[:100]]
y_train = y[data_index[:100]]

#Creating the testing Dataset
#test_randint = randint(0,150,50)
X_test = X[data_index[100:]]
y_test = y[data_index[100:]]

n_neighboures = 7

#Prediction using the function defined in previous cell. 
y_pred = predict(X_train,y_train,X_test , n_neighboures)

#Testing the accuracy of the model
score = accuracy_score(y_test, y_pred)
score
print("n_neighbors : %2d, Accuracy : %5.2f" % (n_neighboures, score))
```
